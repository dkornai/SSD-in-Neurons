```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def diffuse_r(state, rate):
    diff = state[0:-1]*rate
    state[0:-1] -= diff
    state[1:] += diff
    
    return state

def diffuse_l(state, rate):
    diff = state[1:]*rate
    state[1:] -= diff
    state[0:-1] += diff
    
    return state

def dist(n):
    if n == 0:
        return 0
    else:
        return dist(n-1) + np.power(1/2,n-1)
```

```python
state = np.zeros(100); state[10] = 10
distance = np.linspace(0, 10, 100)

mean_prev_pos = np.average(state, weights=distance)
for i in range(500):
    mean_curr_pos = np.average(state, weights=distance)
    #print(np.round(state,2))
    if i % 50 == 0:
        print(mean_prev_pos-mean_curr_pos)
        plt.plot(distance, state)
    #state = diffuse_l(state, 0.01)
    state = diffuse_r(state, 0.1)
    mean_prev_pos = mean_curr_pos
    
```
