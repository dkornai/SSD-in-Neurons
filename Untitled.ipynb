```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def diffuse(state, rate):
    diff = state[0:-1]*rate
    state[0:-1] -= diff
    state[1:] += diff
    
    return state

def dist(n):
    if n == 0:
        return 0
    else:
        return dist(n-1) + np.power(1/2,n-1)
```

```python
state = np.zeros(100); state[0] = 10
distance = [dist(n) for n in range(100)]
for i in range(20):
    #print(np.round(state,2))
    print(np.average(state, weights=distance))
    state = diffuse(state, 0.1)
    
    
plt.scatter(distance, state)
```
